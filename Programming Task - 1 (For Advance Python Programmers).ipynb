```python
def runningSum(nums):
    for i in range(1,len(nums)):
        nums[i]+=nums[i-1]
    return nums
print(runningSum([3,1,2,10,1]))
```

```python
class Solution:
    def shuffle(self,nums,n):
        res=[]
        for i in range(n):
            res.append(nums[i])
            res.append(nums[i+n])
        return res
obj=Solution()
```

```python
res=obj.shuffle([2,5,1,3,4,7],3)
print(res)
```

```python
res=obj.shuffle([1,2,3,4,4,3,2,1],4)
print(res)
```

```python
def kidswithcandies(candies,extracandies):
    max_candies=max(candies)
    res=[]
    for i in range(len(candies)):
        res.append(candies[i]+extracandies>=max_candies)
    return res
print(kidswithcandies([2,3,5,1,3],3))
```
